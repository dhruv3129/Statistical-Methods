```python
list=[6,7,10,12,13,4,8,12]
n=len(list)
list.sort()
if n % 2 == 0:
    median1 = list[n//2]
    median2 = list[n//2 - 1]
    median = (median1 + median2)/2
else:
    median = list[n//2]
print(median)
```

```python
list=[6,7,10,12,13,4,8,12]
n=len(list)
unique = []
mode = []
for i in list:
    if i not in unique:
        unique.append(i)
    else:
        mode.append(i)
print(set(mode))
```

```python
def variance(data):
  n = len(data)
  mean = sum(data) / n
  deviations = [(x - mean) ** 2 for x in data]
  variance = sum(deviations) / n
  return variance
variance([6,7,10,12,13,4,8,12])
```
